```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))

# Generate points for the first ellipse (4x^2 + 9y^2 = 36)
theta1 = np.linspace(0, 2*np.pi, 1000)
a1 = 3
b1 = 2
x1 = a1 * np.cos(theta1)
y1 = b1 * np.sin(theta1)

# Generate points for the second ellipse (x^2 + 3y^2 = 3)
theta2 = np.linspace(0, 2*np.pi, 1000)
a2 = np.sqrt(3)
b2 = 1
x2 = a2 * np.cos(theta2)
y2 = b2 * np.sin(theta2)

# Plot circles
radii = [1, 2, 2.88]
center = (3, 3)  # Center of the circles

for r in radii:
    theta_circle = np.linspace(0, 2*np.pi, 1000)
    x_circle = center[0] + r * np.cos(theta_circle)
    y_circle = center[1] + r * np.sin(theta_circle)
    plt.plot(x_circle, y_circle, label=f'Circle (radius = {r})')

plt.scatter(1.3, 0.62, color='black', label='minimizer')

# Plotting
plt.plot(x1, y1, label='$4x1^2 + 9x2^2 =< 36$')
plt.plot(x2, y2, label='$x1^2 + 32^2 = 3$')
plt.axvline(x=1, color='red', linestyle='--', label='x1 =< 1')  # Vertical line at x = 1
plt.xlabel('x1')
plt.ylabel('x2')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')  # Equal aspect ratio
plt.legend()
plt.show()

```
