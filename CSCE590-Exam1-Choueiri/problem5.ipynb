```python
# Create Plot for Exercise 3 figure
plt.figure(figsize=(8, 6))

plt.axvline(x=3, label='x1=3', color='green')
plt.axvline(x=10, label='x1=10', color='blue')
plt.axhline(y=10, label='x2=10', color='orange')
plt.axhline(y=50, label='x2=50', color='violet')
plt.plot([3, 7], [30, 10], label='10x1+2x2=90', color='red')

x1_vert = [3, 3, 7]
x2_vert = [10, 30, 10]
plt.fill(x1_vert, x2_vert, 'grey')

plt.title('Feasible Region - graphical method')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
```
