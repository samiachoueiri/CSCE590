Promlem 2 

Known : x0 = 0

Decision variables:  x1 , x2

Constrains: 4x2 - 2x1 = 2 ; x2 - x1 = -10 ; x2 + x1 = 2 ; 4x2 + 2x1 = 1   

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t_true = [-2,-1,0,1,2]
y_true = [2,-10,0,2,1]
plt.scatter(t_true, y_true, color='red', linestyle='-')
plt.plot(t_true, y_true, color='blue', linestyle='-')

plt.xlabel('t')
plt.ylabel('y')
plt.title('True values')
plt.show()
```

```python
import numpy as np

# Given data
t_values = np.array([-2, -1, 0, 1, 2])
y_values = np.array([2, -10, 0, 2, 1])

# Define the objective function
def objective_function(x, t_values, y_values):
    p_values = x[0] + x[1]*t_values + x[2]*(t_values**2)
    return np.sum((y_values - p_values)**2)

# Define the gradient of the objective function
def gradient(x, t_values, y_values):
    p_values = x[0] + x[1]*t_values + x[2]*(t_values**2)
    residuals = y_values - p_values
    grad_x0 = -2 * np.sum(residuals)
    grad_x1 = -2 * np.sum(residuals * t_values)
    grad_x2 = -2 * np.sum(residuals * (t_values**2))
    return np.array([grad_x0, grad_x1, grad_x2])

# Gradient descent algorithm
def gradient_descent(t_values, y_values, learning_rate=0.001, max_iter=10000, tol=1e-6):
    # Initialize coefficients
    x = np.zeros(3)
    # Perform gradient descent
    for i in range(max_iter):
        grad = gradient(x, t_values, y_values)
        x -= learning_rate * grad
        # Check convergence
        if np.linalg.norm(grad) < tol:
            break
    return x

# Run gradient descent
coefficients = gradient_descent(t_values, y_values)
# Extract coefficients
x0, x1, x2 = coefficients

# Print the quadratic polynomial
print("Quadratic polynomial coefficients:")
print("x0:", x0)
print("x1:", x1)
print("x2:", x2)
```

```python
t_true = [-2,-1,0,1,2]
y_true = [2,-10,0,2,1]
plt.scatter(t_true, y_true, color='red', linestyle='-')

t = np.linspace(-2, 2, 100)
y = x0 + x1*t + x2*(t**2)
plt.plot(t, y, color='blue', linestyle='-')

plt.xlabel('t')
plt.ylabel('y')
plt.title('Best fit function')
plt.show()
```
